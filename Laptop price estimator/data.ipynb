```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# Define the target URL
url = 'https://www.olx.com.pk/items/q-lenovo-laptop'
pagination_count = 4
```

```python
links = []

# while len(links) == 0:
for page in range(1, pagination_count):
    # Step 2: Fetch the page content
    response = requests.get(url + f'?page={page}')
    print("Status Code:", response.status_code, response)
    soup = BeautifulSoup(response.content, 'html.parser')
    items = soup.find_all('li', {'aria-label': 'Listing'})
    # for each item in the items list extract the a tag and then extract the href attribute
    page_links = [item.find('a')['href'] for item in items]
    links.extend(page_links)
    print(f"Page {page} Done!")

# Remove duplicate links
links = list(set(links))
print(f"Total Unique Links: {len(links)}")

print(links)
```

```python
df = pd.DataFrame(columns=['title', 'price', 'details'])

for link in links:
	link = 'https://www.olx.com.pk' + link
	response = requests.get(link)
	soup = BeautifulSoup(response.content, 'html.parser')
	# find the div with aria-label as 'Overview' and then extract the h1, span tag text
	overview = soup.find('div', {'aria-label': 'Overview'})
	description = soup.find('div', {'aria-label': 'Description'})
	# print(overview.text, description.text)

	
	title = overview.find('h1').text
	price = overview.find('span').text
	details = description.find('span').text

	df.loc[len(df)] = [title, price, details]
```
